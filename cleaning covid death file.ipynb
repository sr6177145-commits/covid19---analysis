```python
import pandas as pd 
import numpy as np
# now we will clean covid death file

#import the file we created  
covid_death=pd.read_csv(r'C:\Users\sr617\Desktop\covid_death.csv')



 
# and some rows where no death or cases there so fill it with 0  

number_col=covid_death.select_dtypes(include='number').columns

covid_death[number_col]=covid_death[number_col].fillna(0)

        


# select the desire column we want to work on 

my_project=covid_death[['iso_code','continent','location','date','new_cases','total_cases','new_deaths','total_deaths','population']]

# set the date in the file
        
my_project['date']=pd.to_datetime(my_project['date'])

my_project.to_csv(r'C:\Users\sr617\Desktop\covid_death_cleaned.csv',index=False)
        

```
